```python
q1 = df['Annual Income (k$)'].quantile(0.25)
q3 = df['Annual Income (k$)'].quantile(0.75)

print("Q1 = ",q1)
print("Q3 = ",q3)
```

```python
IQR = q3-q1
upper_limit = q3 + (1.5)*IQR
lower_limit= q1 - (1.5)*IQR

print("IQR = ",IQR)
print("Upper Limit = ",upper_limit)
print("Lower Limit = ",lower_limit)
```

```python
df=df[(df['Annual Income (k$)']<upper_limit) & (df['Annual Income (k$)']>lower_limit)]
```

# Machine Learning approach with clustering algorithm

```python
from sklearn import cluster
```

```python
error=[]
for i in range(1,11):
  kmeans = cluster.KMeans(n_clusters=i,init = 'k-means++',random_state=42)
  kmeans.fit(new_df)
  error.append(kmeans.inertia_)
```

```python
error
```

```python
plt.plot(range(1,11),error,'b-o')
plt.title('Elbow method')
plt.xlabel('number of clusters')
plt.ylabel('error')
plt.show()
```

```python
km_model = cluster.KMeans(n_clusters = 5,init = 'k-means++',random_state=42)
```

```python
km_model.fit(new_df)
```

```python
pred = km_model.predict(new_df)
pred
```
